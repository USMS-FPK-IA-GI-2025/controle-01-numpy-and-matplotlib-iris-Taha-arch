# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.

# Partie D — Visualisation avec Matplotlib (Plots Simples)

Scatter plot : longueur vs largeur des pétales.


```python

plt.figure(figsize=(8, 6))
plt.scatter(features[:, 2], features[:, 3], alpha=0.6)
plt.xlabel('Longueur des petales')
plt.ylabel('Largeur des petales')
plt.title('Longueur vs largeur')
plt.grid(True)
plt.show()
```

### D19. Bar chart

# Partie E — Visualisations Avancées

### E22. Subplots comparatifs
Deux scatter plots côte à côte : Setosa vs Versicolor.


```python

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

#Setosa
axes[0].scatter(features[:50, 2], features[:50, 3], color='red', alpha=0.6)
axes[0].set_title('Setosa')
axes[0].set_xlabel('Longueur des petales')
axes[0].set_ylabel('Largeur des petales')

#Versicolor
axes[1].scatter(features[50:100, 2], features[50:100, 3], color='blue', alpha=0.6)
axes[1].set_title('Versicolor')
axes[1].set_xlabel('Longueur des petales')
axes[1].set_ylabel('Largeur des petales')

plt.tight_layout()
plt.show()
```

### E23. Grille 2x2
Histogrammes des 4 caractéristiques avec `plt.subplots(2,2)`.


```python

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

labels = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

for i, ax in enumerate(axes.flat):
    ax.hist(features[:, i], bins=15, color='lightblue', edgecolor='black')
    ax.set_title(labels[i])
    ax.set_xlabel('Valeurs (cm)')
    ax.set_ylabel('Frequence')

plt.tight_layout()
plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python

species_data = [features[:50], features[50:100], features[100:]]
means = [np.mean(sp, axis=0) for sp in species_data]
stds_per_species = [np.std(sp, axis=0) for sp in species_data]

x = np.arange(4)
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - width, means[0], width, yerr=stds_per_species[0], label='Setosa')
ax.bar(x, means[1], width, yerr=stds_per_species[1], label='Versicolor')
ax.bar(x + width, means[2], width, yerr=stds_per_species[2], label='Virginica')

ax.set_xticks(x)
ax.set_xticklabels(['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
ax.legend()
plt.show()
```

### E25. Scatter coloré par espèce
Scatter plot avec couleurs par espèce.


```python

plt.figure(figsize=(10, 6))
plt.scatter(features[:50, 2], features[:50, 3], color='red', label='Setosa', alpha=0.6)
plt.scatter(features[50:100, 2], features[50:100, 3], color='blue', label='Versicolor', alpha=0.6)
plt.scatter(features[100:, 2], features[100:, 3], color='green', label='Virginica', alpha=0.6)
plt.xlabel('Longueur des petales')
plt.ylabel('Largeur des petales')
plt.legend()
plt.title('Dimensions des petals par espece')
plt.show()
```

# Partie F — Mini-Analyse Finale

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python

# .T pour transposer le tableau features
correlation_matrix = np.corrcoef(features.T)
print(correlation_matrix)
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python

features_labels = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
correlations = correlation_matrix[2, :]

plt.figure(figsize=(8, 6))
plt.bar(features_labels, correlations, color=['red', 'blue', 'green', 'orange'])
plt.ylabel('Correlation avec Petal Length')
plt.title('Feature Correlations')
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
species_data = [features[:50], features[50:100], features[100:]]
variances = [np.var(sp) for sp in species_data]

print(f"Variance de Setosa:  {variances[0]:.2f}")
print(f"Variance de Versicolor:  {variances[1]:.2f}")
print(f"Variance de Virginica:  {variances[2]:.2f}")

max_variance = np.argmax(variances)
labels = ['Setosa', 'versicolor', 'Virginica']
print(f"L'especes avec la plus grande variance globale: {labels[max_variance]}")
print("\n")

# La visualisation avec matplotlib

max_species_data = species_data[max_variance]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for i, axe in enumerate(axes.flat):
    axe.hist(max_species_data[:, i], bins=10)
    axe.set_title(f'{features_labels[i]} - {labels[max_variance]}')
plt.tight_layout()
plt.show()
```
